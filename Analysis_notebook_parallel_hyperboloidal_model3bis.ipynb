```python
def WE_gaussian_solution(R,T,parameters):
    A = parameters[0]
    ds = parameters[1]
    J = R-T
    L = (R+T)
    P = (A*np.exp(-ds**2*J**2)*J+A*np.exp(-ds**2*L**2)*L)/(2*R)
    return(P)
```

```python
def model1_gaussian_solution(R,T,parameters):
    A = parameters[0]
    ds = parameters[1]
    C = parameters[2]
    J = R-T
    L = (R+T)
    P = (A*np.exp(-ds**2*J**2)*J+A*np.exp(-ds**2*L**2)*L)/(2*R)
    Q = np.log(1+C*P)/C
    return(Q)
```

```python
def model3_gaussian_solution(R,T,parameters):
    A = parameters[0]
    ds = parameters[1]
    C = parameters[2]
    J = R-T
    L = (R+T)
    Chi = (R*R+1.)**0.5
    P = (A*np.exp(-ds**2*J**2)*J+A*np.exp(-ds**2*L**2)*L)/R/2
    Q = np.log(1+P)/C
    return(C*np.sin(Q))
```

```python
def model3bis_gaussian_solution(R,T,parameters):
    A = parameters[0]
    C = parameters[1]
    J = -R+T
    L = (R+T)
    Chi = (R*R+1)**0.5
    P = (+A*np.exp(-ds**2*J**2)*J+A*np.exp(-ds**2*L**2)*L)/R/2
    Q = np.log(1+P*(R*R+1)**0.5)/C
    return(C*np.sin(Q)/(R*R+1)**0.5)
```

```python
def modeldumb_gaussian_solution(R,T,parameters):
    A = parameters[0]
    C = parameters[1]
    J = R-T
    L = (R+T)
    P = (A*np.exp(-ds**2*J**2)*J+A*np.exp(-ds**2*L**2)*L)/2
    Q = (-1/(1+P)+1)/C
    return(C*np.cos(Q))
```

```python
def theo_temporal_vector(X,times,theo_function,parameters):
    temporal_field = []
    for t in times:
        space_set = []
        for x in X:
            space_set.append(theo_function(x,t,parameters))
        temporal_field.append(space_set)
    return(temporal_field)
```

```python
def animate_multiple_fields_theo(vector_of_functions,parameters_vector,time_line,domain,ylim_inf,ylim_sup,functions_name,colors):
    
    data_name = "./data"+str(data[0])+"/fields.mp4"
    fig = plt.figure()
    ax1 = plt.axes(ylim=(ylim_inf,ylim_sup),xlim=(domain[0],domain[-1]))
    line, = ax1.plot([], [])
    plt.xlabel('R')
    plt.ylabel('fields')
    plotlays = [vector_of_functions]
    lines = []
    for i,index in enumerate (functions_name):
        lobj = ax1.plot([],[],c=colors[i],lw=5,alpha=0.6,label=index)[0]
        lines.append(lobj)

    x = []
    y = []
    for i in range(np.size(vector_of_functions)):
        x.append(domain),
        y.append(vector_of_functions[i](domain,0,parameters_vector[i]))


    def init():
        for line in lines:
            line.set_data(x,y)      
            #line.set_data([x1,x2,x3,x4,x5,x6,x7],[y1,y2,y3,y4,y5,y6,y7])
        return lines

    def animate(i):
        for n,j in enumerate (vector_of_functions):
            x[n] = domain
            y[n] = vector_of_functions[n](domain,time_line[i],parameters_vector[n])
        xlist = x
        ylist = y

        #for index in range(0,1):
        for lnum,line in enumerate(lines):
            line.set_data(xlist[lnum], ylist[lnum]) # set data for each line separately. 
        
        return lines

    # call the animator.  blit=True means only re-draw the parts that have changed.
    anim = animation.FuncAnimation(fig, animate, init_func=init,
                                   frames=int(np.size(time_line)), interval=75, blit=True)
    plt.hlines(-1./parameters_vector[1][2],domain[0],domain[-1])
    plt.legend()
    plt.grid()
    plt.show()
    anim.save(data_name)
    
```
